```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv('./data/train.csv')
train.head()
```

```python
train_y = train.Transported.apply(lambda t: 1 if t == True else 0)
train = train.assign(Transported=train_y)
```

```python
vip = train.VIP.apply(lambda t: 1 if t == True else 0)
cryo = train.CryoSleep.apply(lambda t: 1 if t == True else 0)
train = train.assign(VIP=vip,CryoSleep=cryo)
```

```python
train.head()
```

```python
def split_cabin(s):
    return str(s).split('/')[0]
```

```python
cabin = train.Cabin.apply(split_cabin)
train = train.assign(Cabin=cabin)
```

```python
train = pd.get_dummies(train, columns=['HomePlanet', 'Cabin'])
```
